```python
import cv2 as cv2
```

```python
src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
# if src None:
#   print('Image load failed!')
```

```python
src.shape
```

```python
src+=50
```

```python
cv2.imshow('src', src)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src
```

```python
dst = cv2.add(src, 100)
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
for sigma in range(1, 6):
    blurred = cv2.GaussianBlur(src, (0, 0), sigma)

    alpha = 1.0
    dst = cv2.addWeighted(src, 1 + alpha, blurred, -alpha, 0.0)

    desc = "sigma: %d" % sigma
    cv2.putText(dst, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
               1.0, 255, 1, cv2.LINE_AA)

    cv2.imshow('dst', dst)
    cv2.waitKey()
```

```python
snow_gray=cv2.imread('snow.jpg', flags=cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow('snow_gray', snow_gray)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
snow_dst=cv2.add(snow_gray, 100)
```

```python
for sigma in range(1, 10):
    blurred = cv2.GaussianBlur(snow_gray, (0,0), sigma)

    alpha = 1.0
    snow_dst = cv2.addWeighted(snow_gray, 1 + alpha, blurred, -alpha, 0.0)

    desc = "sigma: %d" % sigma
    cv2.putText(snow_dst, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
               1.0, 0, 1, cv2.LINE_AA)

    cv2.imshow('snow_dst', snow_dst)
    cv2.waitKey()
```

```python
snow_gray1=cv2.imread('snow1.jpg', flags=cv2.IMREAD_GRAYSCALE)
```

```python
cv2.imshow('snow_gray1', snow_gray1)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
snow_dst1=cv2.add(snow_gray1, 100)
```

```python
# 1.  (Unsharp Masking)   
# sigma:  .       .
# alpha:  .     .
```

```python
sigma = 3.0
alpah=2.0
for sigma in range(1, 6):
    blurred = cv2.GaussianBlur(snow_gray1, (0,0), sigma)

    alpha = 1.3
    beta = 20
    snow_dst1 = cv2.addWeighted(snow_gray1, 1 + alpha, blurred, -alpha, 0)

    desc = "sigma: %d" % sigma
    cv2.putText(snow_dst1, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
               1.0, (255, 255, 255), 2, cv2.LINE_AA)

    vivid_image = cv2.convertScaleAbs(snow_dst1, alpha=alpha, beta=beta)
    cv2.imshow('snow_dst1', vivid_image)
    cv2.waitKey()
```
